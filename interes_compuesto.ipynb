```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Inicializar variables
años = 30 # Años en los que se invierte
interes = 0.08 # Porcentage de interes generado por el activo
capital_inicial = 1000.0 # Capital inicial de la inversión
capital_mensual = 500.0 # Capital que se añade de manera periodica

interes_compuesto = 0.0 # Inicialización del acumulativo
contribucion = 0.0 # Acumulacion de la contribucion mensual
lista_interes_compuesto = [] # Lista que acumula el beneficio en interes de cada año
lista_beneficios = [] # Es una lista que acumula los beneficios de cada año
lista_contribucion = [] # Es una lista que contiene la acumulacion del capital mensual
total = 0.0
```

```python
for año in range(1, años+1):
    if año == 1:
        total = capital_inicial
        # print("Interés del fondo:", interes)
        # print(f"Capital inicial: {capital_inicial}")

    if capital_mensual > 0:
        for mes in range(1, 12):
            total += capital_mensual
            contribucion += capital_mensual
            lista_contribucion.append(contribucion)
            # print(f"Mes {mes}, Total {total}")

    interes_compuesto = total * interes
    lista_interes_compuesto.append(interes_compuesto)
    total += interes_compuesto
    lista_beneficios.append(total)
    # print(f"Año {año}, Interes compuesto: {interes_compuesto}%")
    # print(f"Fin del ejercicio: {total}€" )
```

```python
xs = np.arange(1, años+1, 1)
ys = lista_beneficios
plt.plot(xs,ys, label='Beneficios')
plt.plot(xs, lista_interes_compuesto, label='Interés')
plt.title("Beneficios respecto del tiempo")
plt.legend()
plt.grid()

print(f"Tienes al final: {total} € habiendo invertido en total {contribucion} €")
```
